```python
#@title Load dataset
import pandas as pd
csv_url = "https://raw.githubusercontent.com/YgZINTJ/ISYS2001_finalExam/refs/heads/main/seasonal_weather_data_exam.csv"
df = pd.read_csv(csv_url)
print(df.info()) # to view data type of each variable, quite helpful
df.head(5)
```

```python
#@title Answer 1: Functions and Control Structures: Coding

##########################################################################
# The requirments are unclear, hence, I have no clue if this is meant to #
# be user input or just adding a conditional column to the original      #
# dataframe, therefore, I choose the latter.                             #
##########################################################################

wind_codition = []

def get_wind(wind_speed):
  if wind_speed < 5:
    result = "Low"
    wind_codition.append(result)

  elif wind_speed >= 5 and wind_speed <= 20:
    result = "Moderate"
    wind_codition.append(result)

  elif wind_speed >20:
    result = "Strong"
    wind_codition.append(result)

for i in df['Wind Speed']:
  get_wind(i)

# Could've use df['Wind Condition'], but I want it to be next to the wind speed.
df.insert(4, "Wind Condition",wind_codition, True)

df.head(5)
```

```python
#@title Answer 2: Data Visualisation with Matplotlib preparing

import matplotlib.pyplot as plt
from datetime import datetime

# user input section
while True:
    try:
        inputMonth = int(input('Enter a Valid Month: '))
        if inputMonth > 12 or inputMonth < 1:
            raise ValueError
    except ValueError:
        print("Be serious, this is an exam, enter again!: ")
        continue
    else:
        break

# extracing month and day
month_list = []
day_list = []

for i in df['Date']:
  datee = datetime.strptime(i, '%Y-%m-%d')
  month_list.append(datee.month)
  day_list.append('Day'+ str(datee.day))

df['Month'] = month_list
df['Days'] = day_list

# function for calculating rain height & setting bar colors in accordence
rain_list = []

def detectRain(hight):
  if hight == 0:
    result = 'No Rain'
    rain_list.append(result)
  elif hight < 5:
    result = 'Light Rain'
    rain_list.append(result)
  elif hight >= 5:
    result = 'Heavy Rain'
    rain_list.append(result)
for j in df['Precipitation']:
  detectRain(j)

df['Rain Height'] = rain_list

colorway = []

def detectColor(hight):
  if hight == 0:
    color = 'Green'
    colorway.append(color)
  elif hight < 5:
    color = 'Yellow'
    colorway.append(color)
  elif hight >= 5:
    color = 'Red'
    colorway.append(color)

for j in df['Precipitation']:
  detectColor(j)

df['color'] = colorway

#
def barPlot (month):
  df_temp = df.loc[df['Month'] == month]
  plt.bar(df_temp['Days'], df_temp['Precipitation'],color = df_temp['color'])
  plt.title('Precipitation in Month ' + str(month))
  plt.xlabel('Days')
  plt.ylabel('Precipitation')
  plt.xticks(rotation = 75)
  plt.show()

def pieChart(month):
  df_temp = df.loc[df['Month'] == month]
  value_count = df_temp['Rain Height'].value_counts()
  df_value_counts = value_count.reset_index()
  df_value_counts.columns = ['Rain Height', 'Count']
  plt.pie(value_count,
          labels = df_value_counts['Rain Height'],
          autopct ='%1.0f%%',
          radius = 1.0)
  plt.title('Precipitation in Month ' + str(month))
  plt.show()

barPlot(inputMonth)
pieChart(inputMonth)

```
